# MODELSKA ANALIZA 1 - 1. naloga

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
from matplotlib.colors import Normalize
from scipy.optimize import curve_fit

plt.rcParams.update({'font.size': 22})
```

## Prvi del naloge

```python
### NU (TAU) ###

def nu(tau, nu0):
    return nu0 + 3/2 * (1-nu0) * (2*tau - tau**2)

nu0s = np.linspace(0, 4, 1000)
taus = np.linspace(0, 1, 1000)

# Create a colormap and normalize based on nu0 values
cmap = get_cmap('tab20')
norm = Normalize(vmin=min(nu0s), vmax=3)

plt.figure(figsize=(12,6))
plt.title(r'$\nu(\tau)$')
plt.ylabel(r'$\nu$')
plt.xlabel(r'$\tau$')
plt.grid()

for nu0 in nu0s:
    nus = nu(taus, nu0)
    # Map nu0 values to colors using the colormap
    if nu0 <= 3:
        color = cmap(norm(nu0))
        plt.plot(taus, nus, color=color)
    else:
        color = 'gray'
        plt.plot(taus, nus, color=color, alpha=0.1)

# Add a colorbar to show the mapping of nu0 values to colors
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, label=r'$\nu_0$')

plt.show()
```

```python
### CHI(TAU) ###

def chi(tau, nu0):
    return nu0*tau + (1-nu0)/2 * (3 * tau**2 - tau**3)

nu0s = np.linspace(0, 4, 1000)
taus = np.linspace(0, 1, 1000)

# Create a colormap and normalize based on nu0 values
cmap = get_cmap('tab20')
norm = Normalize(vmin=min(nu0s), vmax=3)

plt.figure(figsize=(12,6))
plt.title(r'$\chi(\tau)$')
plt.ylabel(r'$\chi$')
plt.xlabel(r'$\tau$')
plt.grid()

for nu0 in nu0s:
    chis = chi(taus, nu0)
    # Map nu0 values to colors using the colormap
    if nu0 <= 3:
        color = cmap(norm(nu0))
        plt.plot(taus, chis, color=color)
    else:
        color = 'gray'
        plt.plot(taus, chis, color=color, alpha=0.1)

# Add a colorbar to show the mapping of nu0 values to colors
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, label=r'$\nu_0$')

plt.show()
```

```python
### Fazni prostor ###

def phaseSpace(tau, nu0):
    return (nu0*tau + (1-nu0)/2 * (3 * tau**2 - tau**3), nu0 + 3/2 * (1-nu0) * (2*tau - tau**2))

nu0s = np.linspace(0, 3, 13)
taus = np.linspace(0, 1, 1000)

# Create a colormap and normalize based on nu0 values
cmap = get_cmap('tab20')
norm = Normalize(vmin=min(nu0s), vmax=3)

plt.figure(figsize=(12,6))
plt.title(r'$\nu(\chi)$')
plt.ylabel(r'$\nu$')
plt.xlabel(r'$\chi$')
plt.grid()

for nu0 in nu0s:
    phase = phaseSpace(taus, nu0)
    # Map nu0 values to colors using the colormap
    if nu0 <= 3:
        color = cmap(norm(nu0))
        plt.plot(phase[0], phase[1], color=color)
    else:
        color = 'gray'
        plt.plot(phase[0], phase[1], color=color, alpha=0.1)

# Add a colorbar to show the mapping of nu0 values to colors
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, label=r'$\nu_0$')

plt.show()
```

## Drugi del naloge (radar)

```python
numax = 1

def nu(tau, nu0, numax):
    return 3 * (2 - nu0 - numax) * (tau - tau**2) + (numax - nu0) * tau + nu0

nu0s = np.linspace(0, 4, 1000)
taus = np.linspace(0, 1, 1000)

# Create a colormap and normalize based on nu0 values
cmap = get_cmap('tab20')
norm = Normalize(vmin=min(nu0s), vmax=4)

plt.figure(figsize=(12,6))
plt.title(r'$\nu(\tau)$')
plt.ylabel(r'$\nu$')
plt.xlabel(r'$\tau$')
plt.grid()

for nu0 in nu0s:
    nus = nu(taus, nu0, numax)
    # Map nu0 values to colors using the colormap
    if nu0 <= 4:
        color = cmap(norm(nu0))
    else:
        color = 'gray'
    plt.plot(taus, nus, color=color)

# Add a colorbar to show the mapping of nu0 values to colors
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, label=r'$\nu_0$')

plt.show()
```

```python
numax = 1

def chi(tau, nu0, numax):
    return (2-nu0-numax) * (3/2 * tau**2 - tau**3) + (numax - nu0)/2 * tau**2 + nu0*tau

nu0s = np.linspace(0, 4, 1000)
taus = np.linspace(0, 1, 1000)

# Create a colormap and normalize based on nu0 values
cmap = get_cmap('tab20')
norm = Normalize(vmin=min(nu0s), vmax=4)

plt.figure(figsize=(12,6))
plt.title(r'$\chi(\tau)$')
plt.ylabel(r'$\chi$')
plt.xlabel(r'$\tau$')
plt.grid()

for nu0 in nu0s:
    nus = chi(taus, nu0, numax)
    # Map nu0 values to colors using the colormap
    if nu0 <= 4:
        color = cmap(norm(nu0))
    else:
        color = 'gray'
    plt.plot(taus, nus, color=color)

# Add a colorbar to show the mapping of nu0 values to colors
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, label=r'$\nu_0$')

plt.show()
```

```python
### Fazni prostor ###

numax = 1

def phaseSpace(tau, nu0):
    return ((2-nu0-numax) * (3/2 * tau**2 - tau**3) + (numax - nu0)/2 * tau**2 + nu0*tau, 3 * (2 - nu0 - numax) * (tau - tau**2) + (numax - nu0) * tau + nu0)

nu0s = np.linspace(0, 3, 13)
taus = np.linspace(0, 1, 1000)

# Create a colormap and normalize based on nu0 values
cmap = get_cmap('tab20')
norm = Normalize(vmin=min(nu0s), vmax=3)

plt.figure(figsize=(12,6))
plt.title(r'$\nu(\chi)$')
plt.ylabel(r'$\nu$')
plt.xlabel(r'$\chi$')
plt.grid()

for nu0 in nu0s:
    phase = phaseSpace(taus, nu0)
    # Map nu0 values to colors using the colormap
    if nu0 <= 3:
        color = cmap(norm(nu0))
        plt.plot(phase[0], phase[1], color=color)
    else:
        color = 'gray'
        plt.plot(phase[0], phase[1], color=color, alpha=0.1)

# Add a colorbar to show the mapping of nu0 values to colors
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, label=r'$\nu_0$')

plt.show()
```

## Tretji del naloge (sode potence pospeška)

```python
nu0 = 0.

def nu(tau, nu0, p):
    return nu0 + (4*p-1)/(2*p) * (1-nu0) * (1-(1-tau)**(2*p/(2*p-1)))

ps = np.arange(1, 11)
taus = np.linspace(0, 1, 1000)

# Create a colormap and normalize based on p values
cmap = get_cmap('rainbow')
norm = Normalize(vmin=min(ps), vmax=max(ps))

plt.figure(figsize=(12,6))
plt.title(r'$\nu(\tau)$')
plt.ylabel(r'$\nu$')
plt.xlabel(r'$\tau$')
plt.grid()

for p in ps:
    nus = nu(taus, nu0, p)
    # Map p values to colors using the colormap
    color = cmap(norm(p))
    plt.plot(taus, nus, color=color)

# Add a colorbar to show the mapping of nu0 values to colors
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, label=r'$p$')

plt.show()
```

## Četrti del naloge (hitrost v funkcionalu)

```python
tauc = 10

def nu(tau, nu0, tauc):
    return (1-nu0/tauc * np.tanh(tauc)) / (1-1/tauc * np.tanh(tauc)) * (1 - np.cosh(tauc*(tau-1)) / np.cosh(tauc)) + nu0 * np.cosh(tauc*(tau-1)) / np.cosh(tauc)

nu0s = np.linspace(0, 4, 1000)
taus = np.linspace(0, 1, 1000)

# Create a colormap and normalize based on nu0 values
cmap = get_cmap('tab20')
norm = Normalize(vmin=min(nu0s), vmax=4)

plt.figure(figsize=(12,6))
plt.title(r'$\nu(\tau)$')
plt.ylabel(r'$\nu$')
plt.xlabel(r'$\tau$')
plt.grid()

for nu0 in nu0s:
    nus = nu(taus, nu0, tauc)
    # Map nu0 values to colors using the colormap
    color = cmap(norm(nu0))
    plt.plot(taus, nus, color=color)

# Add a colorbar to show the mapping of nu0 values to colors
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, label=r'$\nu_0$')

plt.show()
```

## Peti del naloge (Zaporedni semaforji) dva zaporedna

```python
### NU (TAU) ###

def nu(tau, nu0):

    N = 3

    a = np.zeros((3*N, 3*N))
    a[0][2] = 1.
    a[1] = [2.,1.,0.,-2.,-1.,0.,0.,0.,0.]
    a[2] = [1/3, 1/2, 1.,0.,0.,0.,0.,0.,0.]
    a[3] = [1., 1., 1., -1., -1., -1., 0., 0., 0.]
    a[4] = [0.,0.,0.,4.,1.,0.,-4.,-1., 0.]
    a[5] = [0.,0.,0.,7/3, 3/2, 1.,0.,0.,0.]
    a[6] = [0.,0.,0.,4.,2.,1.,-4.,-2.,-1.]
    a[7] = [0.,0.,0.,0.,0.,0.,6.,1.,0.]
    a[8] = [0.,0.,0.,0.,0.,0.,19/3, 5/2, 1.]

    b = np.array([nu0, 0.,1.,0.,0.,1.,0.,0.,1.])

    coef = np.linalg.solve(a, b)

    tau1 = tau[tau <= 1]
    tau2 = tau[tau > 1]
    tau2 = tau2[tau2 <= 2]
    tau3 = tau[tau > 2]

    nu1 = coef[0]*tau1**2 + coef[1]*tau1 + coef[2]
    nu2 = coef[3]*tau2**2 + coef[4]*tau2 + coef[5]
    nu3 = coef[6]*tau3**2 + coef[7]*tau3 + coef[8]

    return np.append(np.append(nu1, nu2), nu3)

nu0s = np.linspace(0, 3.5, 1000)
taus = np.linspace(0, 3, 3000)

# Create a colormap and normalize based on nu0 values
cmap = get_cmap('tab20')
norm = Normalize(vmin=min(nu0s), vmax=max(nu0s))

plt.figure(figsize=(12,6))
plt.title(r'$\nu(\tau)$')
plt.ylabel(r'$\nu$')
plt.xlabel(r'$\tau$')
plt.grid()

for nu0 in nu0s:
    nus = nu(taus, nu0)
    # Map nu0 values to colors using the colormap
    color = cmap(norm(nu0))
    plt.plot(taus, nus, color=color)

# Add a colorbar to show the mapping of nu0 values to colors
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, label=r'$\nu_0$')

plt.show()
```
