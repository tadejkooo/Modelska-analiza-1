```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import gaussian, get_window

plt.rcParams.update({'font.size': 12})
```

```python
nx, ny = 512, 512
N_pad = 512
k1 = np.pad(np.loadtxt('data/lena_slike/kernel1.pgm', skiprows=3).reshape((nx, ny)), N_pad, mode='edge')
k2 = np.pad(np.loadtxt('data/lena_slike/kernel2.pgm', skiprows=3).reshape((nx, ny)), N_pad, mode='edge')
k3 = np.pad(np.loadtxt('data/lena_slike/kernel3.pgm', skiprows=3).reshape((nx, ny)), N_pad, mode='edge')
lena_k1_n0 = np.pad(np.loadtxt('data/lena_slike/lena_k1_n0.pgm', skiprows=3).reshape((nx, ny)), N_pad, mode='edge')
lena_k1_n4 = np.pad(np.loadtxt('data/lena_slike/lena_k1_n4.pgm', skiprows=3).reshape((nx, ny)), N_pad, mode='edge')
lena_k1_n8 = np.pad(np.loadtxt('data/lena_slike/lena_k1_n8.pgm', skiprows=3).reshape((nx, ny)), N_pad, mode='edge')
lena_k1_n16 = np.pad(np.loadtxt('data/lena_slike/lena_k1_n16.pgm', skiprows=3).reshape((nx, ny)), N_pad, mode='edge')
lena_k1_nx = np.pad(np.loadtxt('data/lena_slike/lena_k1_nx.pgm', skiprows=3).reshape((nx, ny)), N_pad, mode='edge')
lena_k2_n0 = np.pad(np.loadtxt('data/lena_slike/lena_k2_n0.pgm', skiprows=3).reshape((nx, ny)), N_pad, mode='edge')
lena_k2_n4 = np.pad(np.loadtxt('data/lena_slike/lena_k2_n4.pgm', skiprows=3).reshape((nx, ny)), N_pad, mode='edge')
lena_k2_n8 = np.pad(np.loadtxt('data/lena_slike/lena_k2_n8.pgm', skiprows=3).reshape((nx, ny)), N_pad, mode='edge')
lena_k2_n16 = np.pad(np.loadtxt('data/lena_slike/lena_k2_n16.pgm', skiprows=3).reshape((nx, ny)), N_pad, mode='edge')
lena_k2_nx = np.pad(np.loadtxt('data/lena_slike/lena_k2_nx.pgm', skiprows=3).reshape((nx, ny)), N_pad, mode='edge')
lena_k3_n0 = np.pad(np.loadtxt('data/lena_slike/lena_k3_n0.pgm', skiprows=3).reshape((nx, ny)), N_pad, mode='edge')
lena_k3_n4 = np.pad(np.loadtxt('data/lena_slike/lena_k3_n4.pgm', skiprows=3).reshape((nx, ny)), N_pad, mode='edge')
lena_k3_n8 = np.pad(np.loadtxt('data/lena_slike/lena_k3_n8.pgm', skiprows=3).reshape((nx, ny)), N_pad, mode='edge')
lena_k3_n16 = np.pad(np.loadtxt('data/lena_slike/lena_k3_n16.pgm', skiprows=3).reshape((nx, ny)), N_pad, mode='edge')
lena_k3_nx = np.pad(np.loadtxt('data/lena_slike/lena_k3_nx.pgm', skiprows=3).reshape((nx, ny)), N_pad, mode='edge')
```

```python
window = get_window('hann', nx+2*N_pad)
window = np.outer(window, window)
plt.figure(figsize=(5, 5))
plt.title('Okno Hann')
plt.imshow(window)
```

```python
K1 = np.fft.fft2(k1)
K2 = np.fft.fft2(k2)
K3 = np.fft.fft2(k3)
LENA_k1_n0 = np.fft.fft2(lena_k1_n0*window)
LENA_k1_n4 = np.fft.fft2(lena_k1_n4*window)
LENA_k1_n8 = np.fft.fft2(lena_k1_n8*window)
LENA_k1_n16 = np.fft.fft2(lena_k1_n16*window)
LENA_k1_nx = np.fft.fft2(lena_k1_nx*window)
LENA_k2_n0 = np.fft.fft2(lena_k2_n0*window)
LENA_k2_n4 = np.fft.fft2(lena_k2_n4*window)
LENA_k2_n8 = np.fft.fft2(lena_k2_n8*window)
LENA_k2_n16 = np.fft.fft2(lena_k2_n16*window)
LENA_k2_nx = np.fft.fft2(lena_k2_nx*window)
LENA_k3_n0 = np.fft.fft2(lena_k3_n0*window)
LENA_k3_n4 = np.fft.fft2(lena_k3_n4*window)
LENA_k3_n8 = np.fft.fft2(lena_k3_n8*window)
LENA_k3_n16 = np.fft.fft2(lena_k3_n16*window)
LENA_k3_nx = np.fft.fft2(lena_k3_nx*window)
```

```python
fig = plt.figure(figsize=(12, 5))

ax = plt.subplot(2, 3, 1)
ax.set_title('kernel1.pgm')
ax.set_ylabel('Surova slika')
ax.set_yticks([])
ax.set_xticks([])
ax.imshow(k1[N_pad:-N_pad, N_pad:-N_pad], cmap='Greys_r')
ax = plt.subplot(2, 3, 4)
ax.set_ylabel('Fourierov prostor')
ax.set_yticks([])
ax.set_xticks([])
ax.imshow(np.abs(K1[N_pad:-N_pad, N_pad:-N_pad]), cmap='Greys_r')

ax = plt.subplot(2, 3, 2)
ax.set_title('kernel2.pgm')
ax.set_yticks([])
ax.set_xticks([])
ax.imshow(k2[N_pad:-N_pad, N_pad:-N_pad], cmap='Greys_r')
ax = plt.subplot(2, 3, 5)
ax.set_yticks([])
ax.set_xticks([])
ax.imshow(np.abs(K2[N_pad:-N_pad, N_pad:-N_pad]), cmap='Greys_r')

ax = plt.subplot(2, 3, 3)
ax.set_title('kernel3.pgm')
ax.set_yticks([])
ax.set_xticks([])
ax.imshow(k3[N_pad:-N_pad, N_pad:-N_pad], cmap='Greys_r')
ax = plt.subplot(2, 3, 6)
ax.set_yticks([])
ax.set_xticks([])
ax.imshow(np.abs(K3[N_pad:-N_pad, N_pad:-N_pad]), cmap='Greys_r')
```

```python
# Slike s kernel 1
sum4 = np.random.normal(0, 4., (nx+2*N_pad, ny+2*N_pad))
SUM4 = np.fft.fft2(np.fft.ifftshift(sum4))
sum8 = np.random.normal(0, 8., (nx+2*N_pad, ny+2*N_pad))
SUM8 = np.fft.fft2(np.fft.ifftshift(sum4))
sum16 = np.random.normal(0, 16., (nx+2*N_pad, ny+2*N_pad))
SUM16 = np.fft.fft2(np.fft.ifftshift(sum4))

Phi0 = (np.abs(LENA_k1_n0)**2) / (np.abs(LENA_k1_n0)**2 + 1000000)
Phi4 = (np.abs(LENA_k1_n4)**2) / (np.abs(LENA_k1_n4)**2 + 4*np.average(np.abs(SUM4))**2)
Phi8 = np.abs(LENA_k1_n8)**2 / (np.abs(LENA_k1_n8)**2 + 10*np.average(np.abs(SUM8))**2.)
Phi16 = np.abs(LENA_k1_n16)**2 / (np.abs(LENA_k1_n16)**2 + 80*np.average(np.abs(SUM16))**2.)

u_k1_n0 = np.fft.ifftshift(np.fft.ifft2(LENA_k1_n0*Phi0/(K1)))
u_k1_n4 = np.fft.ifftshift(np.fft.ifft2(LENA_k1_n4*Phi4/K1))
u_k1_n8 = np.fft.ifftshift(np.fft.ifft2(LENA_k1_n8*Phi8/K1))
u_k1_n16 = np.fft.ifftshift(np.fft.ifft2(LENA_k1_n16*Phi16/K1))
```

```python
fig = plt.figure(figsize=(12, 5))

ax = plt.subplot(2, 4, 1)
ax.set_title('lena_k1_n0.pgm')
ax.set_ylabel('Surova slika')
ax.set_yticks([])
ax.set_xticks([])
ax.imshow(lena_k1_n0[N_pad:-N_pad, N_pad:-N_pad], cmap='Greys_r')
ax = plt.subplot(2, 4, 5)
ax.set_ylabel('Dekonvolucija')
ax.set_yticks([])
ax.set_xticks([])
ax.imshow(np.real(u_k1_n0[N_pad:-N_pad, N_pad:-N_pad]), cmap='Greys_r')

ax = plt.subplot(2, 4, 2)
ax.set_title('lena_k1_n4.dat')
ax.set_yticks([])
ax.set_xticks([])
ax.imshow(lena_k1_n4[N_pad:-N_pad, N_pad:-N_pad], cmap='Greys_r')
ax = plt.subplot(2, 4, 6)
ax.set_yticks([])
ax.set_xticks([])
ax.imshow(np.real(u_k1_n4[N_pad:-N_pad, N_pad:-N_pad]), cmap='Greys_r')

ax = plt.subplot(2, 4, 3)
ax.set_title('lena_k1_n8.dat')
ax.set_yticks([])
ax.set_xticks([])
ax.imshow(lena_k1_n8[N_pad:-N_pad, N_pad:-N_pad], cmap='Greys_r')
ax = plt.subplot(2, 4, 7)
ax.set_yticks([])
ax.set_xticks([])
ax.imshow(np.real(u_k1_n8[N_pad:-N_pad, N_pad:-N_pad]), cmap='Greys_r')

ax = plt.subplot(2, 4, 4)
ax.set_title('lena_k1_n16.dat')
ax.set_yticks([])
ax.set_xticks([])
ax.imshow(lena_k1_n16[N_pad:-N_pad, N_pad:-N_pad], cmap='Greys_r')
ax = plt.subplot(2, 4, 8)
ax.set_yticks([])
ax.set_xticks([])
ax.imshow(np.real(u_k1_n16[N_pad:-N_pad, N_pad:-N_pad]), cmap='Greys_r')
```
